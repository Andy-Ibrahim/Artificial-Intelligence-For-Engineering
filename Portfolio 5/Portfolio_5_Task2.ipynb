```python
import json
import os
from PIL import Image, ImageDraw

# Function to extract polygons and labels from a LabelMe JSON file
def load_annotations(annotation_file):
    with open(annotation_file, 'r') as f:
        data = json.load(f)

    annotations = []
    for shape in data['shapes']:
        label = shape['label']
        points = shape['points']
        annotations.append((label, points))

    return annotations

# Example: Load and display polygons for one image
json_file = os.path.join(json_path, '00300-ZED-Left-1622128958.json')
annotations = load_annotations(json_file)
print("Annotations:", annotations)
```

```python
import matplotlib.pyplot as plt

# Load and display one image with annotations (test with the first image)
test_image = test_image_files[0]
test_json = test_json_files[0]

# Load the image
image_path = os.path.join(images_path, test_image)
image = Image.open(image_path)

# Load the annotations
json_file_path = os.path.join(json_path, test_json)
annotations = load_annotations(json_file_path)
print("Annotations for image:", annotations)

# Draw polygons on the image
draw = ImageDraw.Draw(image)
for label, points in annotations:
    polygon = [(point[0], point[1]) for point in points]
    draw.polygon(polygon, outline="red")

# Display the image using matplotlib
plt.figure(figsize=(8, 8))
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

# Train the Mask R-CNN Model

```python
import os
import numpy as np
import json
from PIL import Image
import matplotlib.pyplot as plt
from mrcnn import utils
import mrcnn.model as modellib
from mrcnn.config import Config
from pycocotools.coco import COCO

# Define configuration
class LogConfig(Config):
    NAME = "log"
    IMAGES_PER_GPU = 1
    NUM_CLASSES = 1 + 1  # Background + log
    STEPS_PER_EPOCH = 100
    DETECTION_MIN_CONFIDENCE = 0.9

config = LogConfig()

# Dataset class
class LogDataset(utils.Dataset):
    def load_log(self, dataset_dir, subset):
        self.add_class("log", 1, "log")
        coco_json_path = os.path.join(dataset_dir, f"{subset}.json")
        dataset_dir = os.path.join(dataset_dir, "images")
        coco_dataset = COCO(coco_json_path)
        image_ids = coco_dataset.getImgIds()
        for image_id in image_ids:
            image_info = coco_dataset.loadImgs(image_id)[0]
            image_path = os.path.join(dataset_dir, image_info["file_name"])
            self.add_image(
                "log",
                image_id=image_id,
                path=image_path,
                width=image_info["width"],
                height=image_info["height"],
                annotations=coco_dataset.loadAnns(coco_dataset.getAnnIds(imgIds=image_id))
            )

    def load_mask(self, image_id):
        info = self.image_info[image_id]
        annotations = info["annotations"]
        masks = np.zeros([info["height"], info["width"], len(annotations)], dtype=np.uint8)
        for i, ann in enumerate(annotations):
            mask = np.zeros([info["height"], info["width"]], dtype=np.uint8)
            for seg in ann["segmentation"]:
                x, y = zip(*np.array(seg).reshape(-1, 2))
                rr, cc = np.polyfill(y, x, (info["height"], info["width"]))
                mask[rr, cc] = 1
            masks[:, :, i] = mask
        return masks.astype(np.bool), np.array([1] * masks.shape[-1], dtype=np.int32)

# Load datasets
dataset_train = LogDataset()
dataset_train.load_log("/content/drive/My Drive/Colab Notebooks/log_labelled_json", "train")
dataset_train.prepare()

dataset_val = LogDataset()
dataset_val.load_log("/content/drive/My Drive/Colab Notebooks/log_labelled_json", "val")
dataset_val.prepare()

# Create model in training mode
model = modellib.MaskRCNN(mode="training", config=config, model_dir="/content/drive/My Drive/Colab Notebooks/log_model")

# Load weights pretrained on COCO
model.load_weights("/path/to/mrcnn_coco_weights.h5", by_name=True, exclude=[
    "mrcnn_class_logits", "mrcnn_bbox_fc",
    "mrcnn_bbox", "mrcnn_mask"])

# Train the model
model.train(dataset_train, dataset_val,
            learning_rate=config.LEARNING_RATE,
            epochs=30,
            layers='heads')
```
